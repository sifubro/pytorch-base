```python
def get_pair_count(ids):
    counts = {}

    # Pythonic way to iterate consecutive elements
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair,0) + 1

    #sorted_counts = sorted( ((v,k) for k,v in counts.items()), reverse=True )
    sorted_count = dict(sorted(counts.items(), key=lambda item: item[1], reverse=True))
    return sorted_count
```

```python
def merge_pair(ids, pair, idx):
    '''
    Take a list of `ids` and  `pair` you want to replace, and that `pair` will be replaced with the new index `idx`.
    So iterating through `ids`, and if we find the `pair` we replace it with `idx`
    '''
    # if the pair contains "|" 
    
    # in the list of ints (ids), replace all consecutive occurences of pair with the new token idx
    newids = []
    i = 0
    
    while i < len(ids):
        # checking for equality at the current position with the pair
        # if we are not at the very last position AND the pair matches, replace it
        if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]: # if we find a match
            newids.append(idx)
            i+=2 # skip the entire pair
        else:
            newids.append(ids[i])
            i+=1
    return newids 
```

```python
# new decoding
def decode_id(int):
    return vocab[int]

def encode_char(char):
    return char_to_int[char]

def decode(id_list):
    # given ids (list of integers) --> return Python String
    decoded_str = ""
    return ''.join((decoded_str + vocab[int] for int in id_list))

def encode(dna_ids):
    # Given a list of strings return a list of integers (the tokens)
    # From text --> integers
    # TODO: Does it need to be converted to list??
    dna_ids = list(map(encode_char, dna_ids))

    # do top to bottom merges
    while True:
        # find pair of bytes that we are allowed to merge according to merges dictionary
        # set of possible merge candidates
        stats = get_pair_count(dna_ids)

        # Identify the pair to merge:
        # find the pair that we are going to be merging: i.e. find a key (pair) inside stats that has the lowest index in the merges dictionary
        # iterate over keys (consecutive pairs) of stats dictionary and take the pair that has the minimum index in merges dictionary
        # hence for any pair inside stats, look into merges at what index it has, and get the pair with the min number
        # if a pair doesn't occur in the merges dict then that pair will not be eligible for merging hence we give it "inf" value (will not be min)
        pair = min(stats, key=lambda p: merges.get(p, float("inf")) )

        # if there is nothing to merge then everything returns "inf" and the pair becomes the first element of stats
        if pair not in merges:
            break #nothing else can be merged

        # do the merging
        idx = merges[pair]
        dna_ids = merge_pair(dna_ids, pair, idx)
    return dna_ids
```

```python
# pad to 1024 tokens
def do_padding(dna_seq, sequence_length=512):
    pad_tokens = [0]*(sequence_length - len(dna_seq))
    return np.array(dna_seq + pad_tokens)

def process_pad(data, sequence_length=512):
    new_dataset = []
    for i, dna_seq in enumerate(data):
        if len(dna_seq) < sequence_length:
            padded_seq = do_padding(dna_seq, sequence_length=sequence_length)
            # if len of dna is up to 511 then fill with <EOS> token
            new_dataset.append(padded_seq)
        else:
            start = 0
            window = 511
            while True:
                subset_dna = dna_seq[start:start+window] #up to 511
                if len(subset_dna) < 10: # too short sequence
                    break
                padded_seq = do_padding(subset_dna, sequence_length=sequence_length)
                new_dataset.append(padded_seq)
                start += 384 #there will be overlap
    print("new dataset length", len(new_dataset))
    return new_dataset
```

## `Self Attention Block`

### Using the `wei.maskked_fill(tril==0, float('-inf'))` we ignore information from the `future` !!!
## `1:09:17 of Let's Build GPT`

```python
# Every single token will emit a query and a key
# query: What I am looking for
# key : what do I contain
# value: what I will communicate to you
# affinities (wei) = q * k
```

```python
import torch.nn as nn

class Head(nn.Module):
    '''
    Implements a single head of self-attention
    '''

    def __init__(self, head_size=64, block_size=512, n_embd=384, dropout=0.2, n_heads=6):
        super().__init__()
        # Lookup Embedding Table
        head_size = n_embd // n_heads
        self.key = nn.Linear(n_embd, head_size, bias=False)
        self.query = nn.Linear(n_embd, head_size, bias=False)
        self.value = nn.Linear(n_embd, head_size, bias=False)
        self.register_buffer('tril', torch.tril(torch.ones(block_size, block_size)))

        self.dropout = nn.Dropout(dropout)
        

    def forward(self, x, targets=None):
        B, T, C = x.shape # (1, 512, 384)
        #print(B,T,C)
        k = self.key(x) # (B, T, C)
        q = self.query(x) # (B, T, C)
        #print(k.shape)
        
        # compute attention scores i.e. "affinities"
        wei = q @ k.transpose(-2, -1) * C ** (-0.5) # (B, T, C) @ (B, C, T) -> (B, T, T)
        #print(wei.shape)

        # remove this to communicate with the future
        wei = wei.masked_fill(self.tril[:T, :T] ==0, float('-inf'))  # (B, T, T)
        
        wei = F.softmax(wei, dim=-1)  # (B, T, T)

        # randomly prevent some nodes from communicating
        wei = self.dropout(wei)
        
        v = self.value(x) # (B, T, C)
        out = wei @ v  # (B, T, T) @ (B, T, C) -> (B, T, C)
        return out
```

```python
class MultiHeadAttention(nn.Module):
    '''
    Multiple heads of self-attention in parallel
    '''

    def __init__(self, num_heads=6, head_size=64, dropout=0.2):
        super().__init__()
        self.heads = nn.ModuleList([ Head(head_size) for _ in range(num_heads)])
        self.proj = nn.Linear(head_size * num_heads, n_embd)
        self.dropout = nn.Dropout(dropout)

    def forward(self,x):
        # x = [1, 512, 384]
        out =  torch.cat([ h(x) for h in self.heads ], dim = -1) # concatenate over the channel dimension
        # out = [1, 512, 24576]
        out = self.dropout( self.proj(out) )
        return out
```

```python
class FFNN(nn.Module):

    def __init__(self, n_embd=384, dropout=0.2):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(n_embd, 4 * n_embd),
            nn.ReLU(),
            nn.Linear(4 * n_embd, n_embd),
            nn.Dropout(dropout)
        )

    def forward(self, x):
        return self.net(x)
```

```python
class GPTLanguageModel(nn.Module):


    def __init__(self, output_classes=1, vocab_size=20, n_embd=384, n_head=6, n_layer=5, block_size=512):
        super().__init__()
        # Lookup Embedding Table
        self.token_embedding_table = nn.Embedding(vocab_size, n_embd)
        self.position_embedding_table = nn.Embedding(block_size, n_embd)
        #self.self_attention_heads = MultiHeadAttention(4, n_embd // 4) # i.e. 4 Heads of 8-dimensional self-attenion (instead of 1 with 32 embed dim)
        #self.ffnn = FFNN(n_embd)

        # self.blocks = nn.Sequential(
        #     Block(n_embd, n_head=4),
        #     Block(n_embd, n_head=4),
        #     Block(n_embd, n_head=4),
        # )
        self.blocks = nn.Sequential(*[ Block(n_embd, n_head=n_head) for _ in range(n_layer)])
        
        
        self.layernorm =  nn.LayerNorm(n_embd)
        self.lm_head = nn.Linear(n_embd, output_classes)

    def forward(self, idx, targets=None):
        #print(idx.shape)
        #print(type(idx))
        #print(idx)
        B, T = idx.shape
        #print(B, T)
        
        # idx and targets are both (B,T) tensor of integers
        tok_emb = self.token_embedding_table(idx)  # (B, T, n_embd=C)
        pos_emb = self.position_embedding_table(torch.arange(T, device=device)) # (T,C=n_embd)
        x = tok_emb + pos_emb # (B, T, C=n_embd)

        # communications across token
        # x = self.self_attention_heads(x) # apply 1 head of self-attention (B, T, C)
        # this is on per-token level independently
        # x = self.ffnn(x) # (B, T, C)
        
        x = self.blocks(x) # (B, T, C)
        x = self.layernorm(x) # (B, T, C)
        
        logits = self.lm_head(x)  # (B, T, vocab_size)

        if targets is None:
            loss = None
        else:
            B, T, C = logits.shape
            logits = logits.view(B*T, C)
            targets = targets.view(-1) # -1 = B*T
            loss = F.cross_entropy(logits, targets)
        
        return logits, loss
```

```python
class Block(nn.Module):


    def __init__(self, n_embd=384, n_head=6):
        super().__init__()
        head_size = n_embd // n_head
        self.sa = MultiHeadAttention(n_embd, head_size)
        self.ffwd = FFNN(n_embd)
        
        # per token transformation
        self.ln1 = nn.LayerNorm(n_embd)  #mean and variance is taken over C=32 numbers. So the batch and the time as act batch dimensions
        self.ln2 = nn.LayerNorm(n_embd)

    def forward(self, x):
        x = x + self.sa( self.ln1(x) )
        x = x + self.ffwd( self.ln2(x) )
        return x
```

```python
import torch
import torch.nn as nn

class ConvNet1D(nn.Module):
    def __init__(self, input_size, num_classes):
        super(ConvNet1D, self).__init__()
        
        # Define the convolutional layers
        self.conv1 = nn.Conv1d(in_channels=input_size, out_channels=16, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv1d(in_channels=16, out_channels=32, kernel_size=3, stride=1, padding=1)
        
        # Define the max pooling layer
        self.pool = nn.MaxPool1d(kernel_size=2, stride=2)
        
        # Calculate the input size to the fully connected layer after pooling
        # We'll do a forward pass to calculate this
        self._to_linear = None
        self.forward(torch.rand(1, input_size, 50))  # Assuming input sequence length of 50
        
        # Define the fully connected layers
        self.fc1 = nn.Linear(self._to_linear, 64)
        self.fc2 = nn.Linear(64, num_classes)
        
        # Define activation and dropout
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.5)
        
    def forward(self, x):
        # Apply first convolutional layer
        x = self.relu(self.conv1(x))
        x = self.pool(x)
        
        # Apply second convolutional layer
        x = self.relu(self.conv2(x))
        x = self.pool(x)
        
        # Flatten for fully connected layers
        if self._to_linear is None:
            self._to_linear = x.shape[1] * x.shape[2]
        
        x = x.view(-1, self._to_linear)
        
        # Apply fully connected layers
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        
        return x
```
